```python
import pandas as pd
import numpy as np
url = "https://raw.githubusercontent.com/leonism/sample-superstore/master/data/superstore.csv"
df = pd.read_csv(url)
df
```

```python
df.columns = df.columns.str.replace(" ", "_").str.lower()
df
```

```python
# #Find total sales and average quantity for each region.

df.groupby("region").agg(
    total_sales= ("sales", "sum"),
    avg_quantity = ("quantity", "mean")

)


```

```python
# .For each category, compute the standard deviation and CV of sales.


df.groupby("category").agg(
    std= ("sales", "std"),
    cv = ("sales", lambda x : (x.std()/x.mean())*100)



)
```

```python
# Create a column high_discount = True if discount > average discount for its segment.
df["high_discount"] = df["discount"] > df.groupby("segment")["discount"].transform("mean")
df["high_discount"]
```

```python
# Filter only those sub-categories where total sales > 1000
df[df.groupby("sub-category")["sales"].transform("sum") > 1000]
```
